## Derivatives Of Common Activation Functions

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Sigmoid
The sigmoid activation function takes in real-valued inputs and returns numbers between 0 and 1.

Definition $1 \over 1 + e^{-x} $
Derivative $

```python
def sigmoid(input):
    return 1 / (1 + np.exp(-input))

def sigmoid_prime(input):
    return sigmoid(input) * (1 - sigmoid(input))
```

```python
def tanh(input):
    return (np.exp(input) - np.exp(-input)) / (np.exp(input) + np.exp(-input))

def tanh_prime(input):
    return 1 - np.power(tanh(input), 2)
```

```python
def relu(input):
    return np.where(input > 0, input, 0)

def relu_prime(input):
    return np.where(input > 0, 1, 0)
```

```python
def leaky_relu(input, alpha=0.02):
    return np.where(input > 0, input, alpha * input)

def leaky_relu_prime(input, alpha=0.02):
    return np.where(input > 0, 1, alpha)
```

```python
data = np.arange(-10,10, 0.1)
```

```python
fig = plt.figure(figsize=[15,8])

ax1 = fig.add_subplot(221)
ax1.plot(sigmoid(data), color='b', label="Sigmoid")
ax1.plot(sigmoid_prime(data), color='r', label="Sigmoid Prime")
ax1.legend()

ax2 = fig.add_subplot(222)
ax2.plot(tanh(data), color='b', label="Tanh")
ax2.plot(tanh_prime(data), color='r', label="Tanh Prime")
ax2.legend()

ax3 = fig.add_subplot(223)
ax3.plot(relu(data), color='b', label="ReLU")
ax3.plot(relu_prime(data), color='r', label="ReLU Prime")
ax3.legend()

ax4 = fig.add_subplot(224)
ax4.plot(leaky_relu(data), color='b', label="Leaky ReLU")
ax4.plot(leaky_relu_prime(data), color='r', label="Leaky ReLU Prime")
ax4.legend()
```
